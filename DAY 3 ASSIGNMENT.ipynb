# Question 1 - Use IF ELSE and ELIF to write a program for your Report Card

```python
n = input("Enter Your name - ")                         #Taking name as input from the user
m = int(input("Enter Marks Obtained - "))               #Taking marks as input from the user and then converting it into int

print("\n")

print("NAME = ", n)                                     #Displaying name
print("Marks Obtained = ", m)                           #Displaying marks obtained
if m >=90:                                              #Checking Grades and displaying it 
    print("Grade S")
elif m >=75 and m<90:
    print("Grade A")
elif m >=60 and m<75:
    print("Grade B")
elif m >=50 and m<60:
    print("Grade C")
elif m >=40 and m<50:
    print("Grade D")
else:
    print("Grade F")
```

# Question 2 - Use For Loop to print Prime Numbers in between 1 to 1000

```python
c = 0
for i in range(1,1001):    #Starting loop from 1 to 1001
    c=0                    #Initializing the c to 0 after each iteration
    for j in range(1,i+1): #Starting loop from i to i+1
        if i%j==0:         #Checking the remainder  
            c=c+1          #Counting the factors
    if c==2:               #checking whether the value of c equals to 2 or not i.e its prime or not because prime numbers has
                                 #only 2 factors
        print(i)           #Displaying the prime numbers
           
            
```

# Question 3 - Write a program for printing the tables from 1 to 10                         

```python
for ii in range(1,11):                           #Starting for loop  
    print("Table of ",ii, "\n")                  #Displaying title of Table
    for jj in range(1,11):                       #Starting for loop 
        print(ii," X ",jj, " = ", ii*jj)         #Calculating and displaying table from 1 to 10
        
    print("\n")                                  #Giving a line break after each table
    
```

# Question 4 - Write a program to print X Prime Numbers using While Loop starting from 0 and take the input of X from the user

```python
l=0
h=int(input("Enter Maximum range : "))                    #Taking the maximum range
print("Prime Numbers between 0 and ",h," are :- ")        #Displaying the given range
l=2
while l <= h:                                             #Checking the values of the given range
    f = -1
    k = 2
    while k <= l/2:                                       #Checking whether number is Prime or not
        if l%k ==0:
            f = f + 1
            k = k + 1
        else:
            k = k + 1
    if f ==-1:
        print(l)                                          #Displaying the Prime numbers
        l=l+1
    else:
        l=l+1
```
